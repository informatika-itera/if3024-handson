# Trivia: Mengapa Pengolahan Sinyal Menggunakan Bilangan Complex??

Banyak dari sinyal yang kita temui itu, secara sederhana merupakan kombinasi dari fungsi sinyal `Sin` dan `Cos`. Namun bekerja dengan kedua sinyal tersebut agak `susah` dalam `integral` maupun `differensial`.

## Euler Formula
Leonhard Euler, mengeluarkan sebuah persamaan untuk merepresentasikan sinyal sinusoidal dalam bentuk eksponen.

$$
e^{j\theta} = \cos\theta + j \sin\theta 
$$

## Apa itu Exponensial

Jika kamu sudah lulus kelas Matematika Dasar TPB, seharusnya kamu familier dengan bilangan euler $e^x$, jika kita plot fungsi ini maka ini grafik yang akan di dapatkan. 

```python
import numpy as np
import matplotlib.pyplot as plt

## Generate x value for e^x
x = np.linspace(-2, 2, 100)
y = np.exp(x)

## Plot the graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = e^x')
plt.show()
```

Namun apa yang terjadi, jika kita tambah bilangan imajiner menjadi $e^{ix}$. Secara visual, grafiknya tidak lagi hanya bergerak naik, namun mengalami rotasi dalam bidang kompleks, membentuk lintasan melingkar.

Notes:
> 3Blue1Brown mempunyai video menarik mengenai relasi dari bilangan imajiner dan euler. [Here](https://www.youtube.com/watch?v=v0YEaeIClKY&ab_channel=3Blue1Brown) 

> Bilangan `i` merupakan nilai dari $i = \sqrt(-1)$. 

> Dalam bidang kompleks, **i = 0 (real) + 1i (imajiner)** , perkalian dengan `i` mengakibatkan rotasi $90^o$ Sehingga grafik $e^{ix}$ terlihat seperti lingkaran.  

Sederhananya, karena $e^{ix}$ merepresentasikan rotasi dalam bentuk lingkaran, maka secara tidak langsung kita dapat bekerja dengan konsep trigonometri, yang erat kaitannya dengan lingkaran. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 1000)

# Compute the corresponding real and imaginary parts for e^(ix) and e^(-ix)
real_part = np.cos(x)
imaginary_part = np.sin(x)

# Create the first plot for e^(ix)
fig1 = plt.figure()
ax = fig1.add_subplot(111, projection='3d')
ax.plot(x, real_part, imaginary_part, label='$e^{ix}$')
ax.set_xlabel('x')
ax.set_ylabel('Real Part')
ax.set_zlabel('Imaginary Part')
ax.set_title('3D Visualization of $e^{ix}$')
ax.legend()
ax.view_init(elev=20, azim=60)

# Show the first plot
plt.show()

# Create the second plot for e^(-ix)
fig2 = plt.figure()
bx = fig2.add_subplot(111, projection='3d')
bx.plot(x, real_part, imaginary_part, label='$e^{-ix}$')
bx.set_xlabel('x')
bx.set_ylabel('Real Part')
bx.set_zlabel('Imaginary Part')
bx.set_title('3D Visualization of $e^{-ix}$')
bx.legend()
bx.view_init(elev=0, azim=0)

# Show the second plot
plt.show()
```

## Taylor Series

Taylor series menjelaskan bahwa sebuah fungsi matematika itu dapat dijabarkan dalam bentuk `polynomial`.

![taylor-series](../media/taylor-series.png)

**Notes**:
> Video menarik dari 3Blue1Brown mengenai Taylor Series. [Here](https://www.youtube.com/watch?v=3d6DsjIBzJ4)

Jika kita representasikan fungsi $e^{ix}$, maka ini yang akan kita dapatkan
$$
e^{ix} = 1 + (ix) + \frac{(ix)^2}{2!} + \frac{(ix)^3}{3!} + ...
$$

Karena $i^2 = -1$, maka $i^2 = -1$, $i^3 = -i$, $i^4 = 1$, sehingga 

$$ 
e^{ix} = 1 + (ix) + - \frac{(x)^2}{2!} - i\frac{(x)^3}{3!} + \frac{(x)^4}{4!} + i\frac{(5)^5}{5!} + ...
$$

Bisa kita group persamaan ini menjadi:

$$
e^{ix} = (1 - \frac{(x)^2}{2!} + \frac{(x)^4}{4!} + ...) + i(x - \frac{x^3}{3!} + \frac{5^5}{5!} + ...)
$$

Lihat, kedua bagian itu mirip dengan `Cos` dan `Sin`, sehingga

$$
e^{ix} = \cos(x) + i\sin(x)
$$


### Perfect Equation
Jika kita ganti `x` dengan $\pi$:

$$
e^{i\pi} = \cos(\pi) + i\sin(\pi) \\
e^{i\pi} = - 1 \\
e^{i\pi} + 1 = 0
$$

Ini yang disebut dengan `Euler Identity` atau `Perfect Equation` jika kamu cari di [Google](https://www.google.com/search?q=perfect+equation+in+math).
