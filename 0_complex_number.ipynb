### Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Create a Basic Signals

Letâ€™s create a basic sinusoidal signal, which is a common type of signal in signal processing.



```python
# Time array from 0 to 1 second with 1000 samples
t = np.linspace(0, 1, 1000)

# Frequency of the signal (5 Hz)
frequency = 5

# Create a sinusoidal signal
signal = np.sin(2 * np.pi * frequency * t)

# Plot the signal
plt.figure(figsize=(10, 3))
plt.plot(t, signal)
plt.title('Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### Represent Signal Using Complex Numbers
Now, weâ€™ll represent this sinusoidal signal using complex numbers. Remember that a sinusoid can be written as a combination of cosine and sine functions, which relates directly to the real and imaginary parts of a complex number.

```python
# Create a complex representation of the signal
complex_signal = np.exp(1j * 2 * np.pi * frequency * t)
print(f"First 5 samples of the complex signal: {complex_signal[:5]}")

# Separate into real and imaginary parts
real_part = np.real(complex_signal)
imaginary_part = np.imag(complex_signal)

print(f"First 5 samples of the real part: {real_part[:5]}")
print(f"First 5 samples of the imaginary part: {imaginary_part[:5]}")

# Plot the real and imaginary parts
plt.figure(figsize=(10, 3))
plt.plot(t, real_part, label='Real Part (cosine)')
plt.plot(t, imaginary_part, label='Imaginary Part (sine)')
plt.title('Complex Representation of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Visualize in Polar Form

```python
# Compute magnitude and phase
magnitude = np.abs(complex_signal)
phase = np.angle(complex_signal)

# Plot magnitude and phase
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, magnitude)
plt.title('Magnitude of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, phase)
plt.title('Phase of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Explanation
When you run this code:

1. **Sinusoidal Signal**: The first plot shows a simple sinusoidal wave. This is the basic signal weâ€™re working with.
2. **Complex Representation**: The second set of plots shows how this signal can be broken down into a real part (cosine wave) and an imaginary part (sine wave). This directly relates to \( z = x + jy \).
3. **Magnitude and Phase**: The final plots show the magnitude and phase over time. The magnitude is constant because the signal doesnâ€™t change in amplitude, and the phase changes linearly over time, corresponding to the signal's frequency.

### How This Relates to Signal Processing
In signal processing, engineers often use this complex representation to analyze and manipulate signals. For example:
- **Real and Imaginary Parts**: Can be used to represent the different components of a signal.
- **Magnitude and Phase**: These tell us how strong the signal is and how its timing changes over time.
