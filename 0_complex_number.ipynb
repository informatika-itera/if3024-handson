### Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Create a Basic Signals

Letâ€™s create a basic sinusoidal signal, which is a common type of signal in signal processing.



```python
# Time array from 0 to 1 second with 1000 samples
t = np.linspace(0, 1, 1000)

# Frequency of the signal (5 Hz)
frequency = 5

# Create a sinusoidal signal
signal = np.sin(2 * np.pi * frequency * t)

# Plot the signal
plt.figure(figsize=(10, 3))
plt.plot(t, signal)
plt.title('Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### Represent Signal Using Complex Numbers
Now, weâ€™ll represent this sinusoidal signal using complex numbers. Remember that a sinusoid can be written as a combination of cosine and sine functions, which relates directly to the real and imaginary parts of a complex number.

```python
# Create a complex representation of the signal
complex_signal = np.exp(1j * 2 * np.pi * frequency * t)
print(f"First 5 samples of the complex signal: {complex_signal[:5]}")

# Separate into real and imaginary parts
real_part = np.real(complex_signal)
imaginary_part = np.imag(complex_signal)

print(f"First 5 samples of the real part: {real_part[:5]}")
print(f"First 5 samples of the imaginary part: {imaginary_part[:5]}")

# Plot the real and imaginary parts
plt.figure(figsize=(10, 3))
plt.plot(t, real_part, label='Real Part (cosine)')
plt.plot(t, imaginary_part, label='Imaginary Part (sine)')
plt.title('Complex Representation of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Visualize in Polar Form

```python
# Compute magnitude and phase
magnitude = np.abs(complex_signal)
phase = np.angle(complex_signal)

# Plot magnitude and phase
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, magnitude)
plt.title('Magnitude of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, phase)
plt.title('Phase of the Signal')
plt.xlabel('Time (s)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Explanation
When you run this code:

1. **Sinusoidal Signal**: The first plot shows a simple sinusoidal wave. This is the basic signal weâ€™re working with.
2. **Complex Representation**: The second set of plots shows how this signal can be broken down into a real part (cosine wave) and an imaginary part (sine wave). This directly relates to \( z = x + jy \).
3. **Magnitude and Phase**: The final plots show the magnitude and phase over time. The magnitude is constant because the signal doesnâ€™t change in amplitude, and the phase changes linearly over time, corresponding to the signal's frequency.

### How This Relates to Signal Processing
In signal processing, engineers often use this complex representation to analyze and manipulate signals. For example:
- **Real and Imaginary Parts**: Can be used to represent the different components of a signal.
- **Magnitude and Phase**: These tell us how strong the signal is and how its timing changes over time.

# Signal and It's Properties

### **Pengenalan Membuat Sinyal di Python**

Dalam pemrosesan sinyal digital, **sinyal** adalah fungsi dari waktu (atau ruang) yang menyampaikan informasi. Sinyal dapat dibuat dalam berbagai bentuk, seperti gelombang sinusoidal (sinewave), gelombang persegi (square wave), gelombang gigi gergaji (sawtooth wave), noise acak, dan lainnya. Dalam Python, sinyal sering direpresentasikan sebagai array (vektor) dari nilai numerik yang sesuai dengan amplitudo sinyal pada titik waktu yang berbeda.

### **Konsep Dasar Membuat Sinyal**

1. **Sampling Rate (`fs`)**: Ini adalah jumlah sampel yang diambil per detik dan diukur dalam Hertz (Hz). Sampling rate menentukan resolusi sinyal dalam domain waktu.
   
2. **Vektor Waktu (`t`)**: Ini merepresentasikan titik-titik waktu diskrit di mana sinyal diambil sampelnya. Biasanya merupakan vektor yang terdistribusi secara linier.

3. **Frekuensi (`f`)**: Ini adalah jumlah siklus per detik (Hz) untuk sinyal periodik seperti gelombang sinus.

4. **Amplitudo (`A`)**: Ini adalah nilai maksimum dari sinyal.

### **Metode Umum Membuat Sinyal di Python**

Berikut adalah beberapa cara umum untuk membuat sinyal di Python:

#### 1. **Sinyal Sinusoidal**
Sinyal sinusoidal adalah salah satu jenis sinyal paling dasar yang digunakan dalam pemrosesan sinyal.

- **Formula**: `x(t) = A * sin(2 * Ï€ * f * t + Ï†)`
  - `A`: Amplitudo
  - `f`: Frekuensi (Hz)
  - `Ï†`: Fase (radian)
  - `t`: Vektor waktu

```python
# Parameter
fs = 1000  # Sampling rate (sampel per detik)
t = np.linspace(0, 1, fs, endpoint=False)  # Vektor waktu (durasi 1 detik)
f = 5  # Frekuensi gelombang sinus (Hz)
A = 1  # Amplitudo

# Membuat sinyal sinusoidal
sinusoidal_signal = A * np.sin(2 * np.pi * f * t)

# Plot sinyal
plt.figure(figsize=(15, 3))
plt.plot(t, sinusoidal_signal)
plt.title('Sinyal Sinusoidal')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

#### 2. **Square Wave**
Square wave berganti antara nilai tinggi dan rendah, tetap pada setiap level untuk durasi yang tetap.

```python
  from scipy import signal

  # Membuat square wave
  square_wave = signal.square(2 * np.pi * f * t)

  # Plot sinyal
  plt.figure(figsize=(15, 3))
  plt.plot(t, square_wave)
  plt.title('Sinyal Square Wave')
  plt.xlabel('Waktu [s]')
  plt.ylabel('Amplitudo')
  plt.grid(True)
  plt.show()
```

#### 3. **Sawtooth Wave**
Sawtooth wave adalah bentuk gelombang periodik yang naik ke atas dan turun tajam.

```python
sawtooth_wave = signal.sawtooth(2 * np.pi * f * t)

  # Plot sinyal
plt.figure(figsize=(15, 3))
plt.plot(t, sawtooth_wave)
plt.title('Sinyal Sawtooth Wave')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

#### 4. **Noise Acak**
Noise acak dapat digunakan untuk mensimulasikan lingkungan yang bising atau untuk menguji ketahanan filter dan algoritma.

```python
noise_signal = np.random.normal(0, 1, t.shape)

# Plot sinyal
plt.figure(figsize=(15, 3))
plt.plot(t, noise_signal)
plt.title('Sinyal Noise Acak')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

#### 5. **Sinyal Komposit**
Sinyal komposit adalah kombinasi dari beberapa sinyal, seperti menambahkan beberapa gelombang sinusoidal bersama-sama.

```python
# Membuat sinyal komposit (jumlah dua gelombang sinus)
f1 = 5  # Frekuensi gelombang sinus pertama
f2 = 20  # Frekuensi gelombang sinus kedua

composite_signal = A * np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# Plot sinyal
plt.figure(figsize=(15, 3))
plt.plot(t, composite_signal)
plt.title('Sinyal Komposit')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

### **Properti Sinyal**

- **Amplitudo**: Nilai puncak dari sinyal. Misalnya, dalam gelombang sinusoidal `x(t) = A * sin(2 * Ï€ * f * t)`, `A` adalah amplitudo.
  
- **Frekuensi**: Jumlah siklus per detik (diukur dalam Hz) untuk sinyal periodik seperti gelombang sinus.

- **Fase**: Offset dari gelombang sinus dari titik asalnya, sering direpresentasikan dalam radian. Ini menggeser gelombang ke kiri atau kanan pada sumbu waktu.

- **Sampling Rate (`fs`)**: Jumlah sampel per detik yang digunakan untuk mendigitalkan sinyal. Sampling rate yang lebih tinggi menangkap lebih banyak detail dari sinyal.

- **Periode Waktu (`T`)**: Untuk sinyal periodik, ini adalah kebalikan dari frekuensi (`T = 1/f`). Ini merepresentasikan waktu yang dibutuhkan untuk menyelesaikan satu siklus penuh.

- **Panjang Gelombang**: Untuk sinyal yang merambat dalam ruang, ini adalah jarak di mana bentuk gelombang berulang.

Di Python, sinyal dapat dibuat menggunakan berbagai fungsi dan metode matematika, masing-masing dengan propertinya yang unik:

- **Sawtooth wave** memiliki kenaikan linier dan penurunan tajam.
