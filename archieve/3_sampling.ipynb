## 2.1 Sampling

````python
# Continuous time variable
t_continuous = np.linspace(0, 1, 1000)  # 1000 points between 0 and 1 second

# Continuous signal (sine wave)
f_continuous = 5  # Frequency of the sine wave
y_continuous = np.sin(2 * np.pi * f_continuous * t_continuous)

# Plot the continuous signal
plt.figure(figsize=(10, 4))
plt.plot(t_continuous, y_continuous, label='Continuous Signal')
plt.title('Continuous Signal (Sine Wave)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

```python

```
Bagian ini mendefinisikan variabel waktu kontinu `t_continuous` dengan menggunakan `numpy.linspace`.
- `np.linspace(0, 1, 1000)` adalah fungsi dari `numpy` yang menghasilkan array dengan 1000 titik yang terdistribusi secara merata antara 0 dan 1 detik.
- Variabel `t_continuous` akan digunakan sebagai sumbu waktu untuk sinyal kontinu yang akan dihasilkan.

Variabel waktu kontinu diperlukan untuk mensimulasikan sinyal analog yang berubah terus-menerus dalam domain waktu. Penggunaan 1000 titik memberikan resolusi yang cukup tinggi untuk representasi sinyal yang halus.

```python
# Continuous signal (sine wave)
f_continuous = 5  # Frequency of the sine wave
y_continuous = np.sin(2 * np.pi * f_continuous * t_continuous)
```
Bagian ini menghasilkan sinyal kontinu berupa gelombang sinus.

- `f_continuous = 5` mendefinisikan frekuensi gelombang sinus sebesar 5 Hz.
- `np.sin(2 * np.pi * f_continuous * t_continuous)` menghitung nilai gelombang sinus pada setiap titik waktu dalam `t_continuous`.
  - `2 * np.pi * f_continuous * t_continuous` mengkonversi frekuensi dari Hertz ke radian per detik untuk penggunaan dalam fungsi sinus.

Gelombang sinus adalah sinyal dasar dalam banyak aplikasi teknik dan ilmiah. Frekuensi 5 Hz berarti sinyal tersebut memiliki lima siklus penuh dalam satu detik.

````python
# Sampling rates
sampling_rates = [10, 20, 50]  # 10 Hz, 20 Hz, 50 Hz

plt.figure(figsize=(15, 10))

for i, fs in enumerate(sampling_rates, 1):
    t_sampled = np.arange(0, 1, 1/fs)  # Sampled time points
    y_sampled = np.sin(2 * np.pi * f_continuous * t_sampled)  # Sampled signal
    
    # Plot sampled signal
    plt.subplot(3, 1, i)
    plt.plot(t_continuous, y_continuous, label='Continuous Signal')
    plt.stem(t_sampled, y_sampled, 'r', markerfmt='ro', basefmt='r-', label=f'Sampled Signal ({fs} Hz)')
    plt.title(f'Sampling at {fs} Hz')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
````

## 2.2 Signal Reconstruction

````python
# Continuous time variable
t_continuous = np.linspace(0, 1, 1000)  # 1000 points between 0 and 1 second

# Continuous signal (sine wave)
f_continuous = 5  # Frequency of the sine wave
y_continuous = np.sin(2 * np.pi * f_continuous * t_continuous)

# Plot the continuous signal
plt.figure(figsize=(10, 4))
plt.plot(t_continuous, y_continuous, label='Continuous Signal')
plt.title('Continuous Signal (Sine Wave)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Sampling rates
sampling_rates = [10, 20, 50]  # 10 Hz, 20 Hz, 50 Hz

# Dictionary to hold sampled data
sampled_signals = {}

plt.figure(figsize=(15, 10))

for i, fs in enumerate(sampling_rates, 1):
    t_sampled = np.arange(0, 1, 1/fs)  # Sampled time points
    y_sampled = np.sin(2 * np.pi * f_continuous * t_sampled)  # Sampled signal
    
    # Store sampled data
    sampled_signals[fs] = (t_sampled, y_sampled)
    
    # Plot sampled signal
    plt.subplot(3, 1, i)
    plt.plot(t_continuous, y_continuous, label='Continuous Signal')
    plt.stem(t_sampled, y_sampled, 'r', markerfmt='ro', basefmt='r-', label=f'Sampled Signal ({fs} Hz)')
    plt.title(f'Sampling at {fs} Hz')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
# Reconstruction of signals from sampled data
plt.figure(figsize=(15, 10))

for i, fs in enumerate(sampling_rates, 1):
    t_sampled, y_sampled = sampled_signals[fs]
    
    # Reconstruct the signal using zero-order hold (step function)
    t_reconstructed = np.linspace(0, 1, 1000)  # Time points for reconstruction
    y_reconstructed = np.interp(t_reconstructed, t_sampled, y_sampled)  # Interpolated signal
    
    # Plot reconstructed signal
    plt.subplot(3, 1, i)
    plt.plot(t_continuous, y_continuous, label='Continuous Signal')
    plt.plot(t_reconstructed, y_reconstructed, '--', label=f'Reconstructed Signal ({fs} Hz)')
    plt.title(f'Reconstruction from {fs} Hz Sampled Signal')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
````

- **Interpolasi Linear**: Metode untuk memperkirakan nilai antara dua titik data yang diketahui. Dalam konteks ini, interpolasi linear digunakan untuk mengisi nilai-nilai sinyal di antara titik-titik sampling.

Bagian kode ini penting untuk memahami bagaimana sinyal yang disampling dapat direkonstruksi kembali menjadi bentuk yang lebih kontinu menggunakan teknik interpolasi. Ini menunjukkan bahwa meskipun sinyal asli telah disampling, kita masih dapat merekonstruksinya dengan akurasi yang baik jika laju sampling cukup tinggi.

## 2.3 Resampling / Interpolating

````python
# High sampling rate
high_fs = 1000  # 1000 Hz
t_high = np.linspace(0, 1, high_fs, endpoint=False)  # 1 second worth of data

# High-sampling-rate signal (sine wave)
f_high = 5  # Frequency of the sine wave
y_high = np.sin(2 * np.pi * f_high * t_high)

# Plot the high-sampling-rate signal
plt.figure(figsize=(10, 4))
plt.plot(t_high, y_high, label='High-Sampling-Rate Signal')
plt.title('High-Sampling-Rate Signal (Sine Wave)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Add noise to the high-sampling-rate signal
f_high = 5  # Frequency of the sine wave
noise = np.sin(10 * np.pi * f_high * t_high)
noise2 = np.sin(0.7 * np.pi * f_high * t_high)
y_noisy = y_high + noise + noise2

# Combine with another sine wave
f_combined = 50  # Frequency of the combined sine wave
y_combined = y_noisy + 0.5 * np.sin(2 * np.pi * f_combined * t_high)

# Plot the noisy combined signal
plt.figure(figsize=(10, 4))
plt.plot(t_high, y_combined, label='Noisy Combined Signal')
plt.title('Noisy Combined Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

```python

```
**Deskripsi:**
Bagian ini menambahkan noise ke sinyal dengan laju sampling tinggi (`y_high`).

**Penjelasan:**
- `f_high = 5` menetapkan frekuensi gelombang sinus asli sebesar 5 Hz.
- `noise = np.sin(10 * np.pi * f_high * t_high)` menghasilkan sinyal noise pertama dengan frekuensi 10 kali lipat dari sinyal asli.
- `noise2 = np.sin(0.7 * np.pi * f_high * t_high)` menghasilkan sinyal noise kedua dengan frekuensi 0.7 kali lipat dari sinyal asli.
- `y_noisy = y_high + noise + noise2` menggabungkan sinyal asli `y_high` dengan kedua sinyal noise, menghasilkan sinyal `y_noisy` yang tercampur noise.

**Teori:**
- **Noise**: Sinyal gangguan yang ditambahkan ke sinyal asli untuk mensimulasikan kondisi nyata di mana sinyal sering terganggu oleh berbagai jenis noise.
- **Frekuensi Noise**: Noise dengan frekuensi berbeda digunakan untuk mensimulasikan berbagai jenis gangguan. Noise berfrekuensi tinggi dapat merepresentasikan interferensi atau harmonik yang tidak diinginkan, sementara noise berfrekuensi rendah bisa merepresentasikan drift atau variasi jangka panjang.
- **Superposisi**: Prinsip bahwa beberapa sinyal dapat dijumlahkan untuk menghasilkan sinyal gabungan. Dalam kasus ini, sinyal asli dan noise dijumlahkan untuk menghasilkan sinyal noisy.

```python
# Combine with another sine wave

```
**Deskripsi:**
Bagian ini menggabungkan sinyal noisy dengan sinyal gelombang sinus tambahan.

**Penjelasan:**
- `f_combined = 50` menetapkan frekuensi gelombang sinus tambahan sebesar 50 Hz.
- `y_combined = y_noisy + 0.5 * np.sin(2 * np.pi * f_combined * t_high)` menambahkan sinyal gelombang sinus tambahan dengan amplitudo 0.5 ke sinyal noisy `y_noisy`, menghasilkan sinyal gabungan `y_combined`.

**Teori:**
- **Sinyal Tambahan**: Menambahkan sinyal gelombang sinus lain ke sinyal noisy untuk mensimulasikan situasi di mana beberapa sinyal berinterferensi atau tumpang tindih.
- **Superposisi Sinyal**: Prinsip bahwa sinyal-sinyal dapat dijumlahkan untuk menghasilkan sinyal gabungan. Ini sering terjadi dalam sistem nyata di mana beberapa sumber sinyal mempengaruhi hasil akhir.

Bagian kode ini penting untuk memahami bagaimana sinyal dapat terganggu oleh noise dan bagaimana sinyal tambahan dapat mempengaruhi sinyal utama. Ini mensimulasikan skenario nyata di mana sinyal yang diinginkan sering tercampur dengan noise dan sinyal interferensi lainnya, dan menunjukkan pentingnya teknik pemrosesan sinyal untuk mengekstrak informasi yang diinginkan dari sinyal noisy.

````python
# Lower sampling rate
low_fs = 100  # 100 Hz
t_low = np.linspace(0, 1, low_fs, endpoint=False)  # New time points for resampling

# Resample the signal using numpy interpolation
y_resampled = np.interp(t_low, t_high, y_combined)

# Plot the resampled signal
plt.figure(figsize=(10, 4))
plt.plot(t_high, y_combined, label='Original Noisy Combined Signal')
plt.plot(t_low, y_resampled, label='Resampled Signal (100 Hz)')
plt.title('Resampled Signal (Lower Sampling Rate)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
````

**Teori:**

# Sampled signal

#### Kuantisasi Sinyal yang Disampling

```python
# Quantization parameters
num_bits = 3  # Number of bits for quantization
num_levels = 2**num_bits  # Number of quantization levels
```
**Deskripsi:**
Bagian ini menetapkan parameter-parameter yang digunakan untuk proses kuantisasi sinyal.

**Penjelasan:**
- `num_bits = 3` menetapkan jumlah bit yang digunakan untuk kuantisasi.
- `num_levels = 2**num_bits` menghitung jumlah level kuantisasi yang ada berdasarkan jumlah bit. Dengan 3 bit, ada \(2^3 = 8\) level kuantisasi.

**Teori:**
- **Kuantisasi**: Proses di mana nilai-nilai sinyal diskrit dipetakan ke nilai-nilai yang terbatas dalam domain digital. Ini adalah langkah kedua dalam proses ADC setelah sampling.
- **Jumlah Bit**: Menentukan resolusi kuantisasi. Semakin banyak bit yang digunakan, semakin tinggi resolusi kuantisasi, dan semakin dekat nilai digital dengan nilai asli sinyal.
- **Jumlah Level Kuantisasi**: Ditentukan oleh \(2^{\text{num\_bits}}\), yaitu jumlah nilai diskrit yang dapat diambil oleh sinyal setelah kuantisasi.

```python
# Quantize the sampled signal
y_quantized = np.round((y_sampled + 1) * (num_levels / 2 - 1)) / (num_levels / 2 - 1) - 1
```
**Deskripsi:**
Bagian ini melakukan proses kuantisasi pada sinyal yang telah disampling.

**Penjelasan:**
- `y_quantized = np.round((y_sampled + 1) * (num_levels / 2 - 1)) / (num_levels / 2 - 1) - 1` menghitung nilai kuantisasi untuk setiap sampel dalam `y_sampled`.
  - `(y_sampled + 1) * (num_levels / 2 - 1)` menormalisasi nilai `y_sampled` ke rentang 0 hingga `num_levels - 1`.
  - `np.round()` membulatkan nilai-nilai ini ke level kuantisasi terdekat.
  - `/ (num_levels / 2 - 1) - 1` menormalisasi kembali nilai-nilai yang dibulatkan ke rentang asli -1 hingga 1.

**Teori:**
- **Kuantisasi**: Menormalisasi nilai-nilai sinyal ke dalam rentang yang dapat diwakili oleh level kuantisasi, membulatkan ke nilai terdekat, lalu menormalisasi kembali ke rentang aslinya.
- **Pembulatan**: Fungsi `np.round()` digunakan untuk membulatkan nilai-nilai yang telah dinormalisasi ke level kuantisasi terdekat.
- **Normalisasi**: Proses menyesuaikan nilai-nilai sinyal ke dalam rentang tertentu (misalnya, -1 hingga 1) untuk keperluan kuantisasi dan kemudian mengembalikannya ke rentang aslinya setelah pembulatan.

Bagian kode ini penting untuk memahami langkah kedua dalam proses ADC, yaitu kuantisasi. Dengan mengkuantisasi sinyal yang telah disampling, kita mengubahnya menjadi nilai-nilai diskrit yang dapat diwakili dalam domain digital. Ini adalah langkah kritis dalam mengubah sinyal analog menjadi sinyal digital yang dapat diproses dan dianalisis lebih lanjut.

````python
# Quantization parameters (reused from previous step)
num_bits = 3  # Number of bits for quantization
num_levels = 2**num_bits  # Number of quantization levels

# Define the quantization process and obtain quantized values
quantized_values = np.round((y_sampled + 1) * (num_levels / 2 - 1)) / (num_levels / 2 - 1) - 1

# Map quantized values to binary codes
def quantized_to_binary(value, num_bits):
    # Convert value to an integer index
    index = int((value + 1) * (num_levels / 2 - 1))
    # Convert index to binary with fixed number of bits
    return format(index, f'0{num_bits}b')

# Apply the function to the quantized values
binary_codes = [quantized_to_binary(val, num_bits) for val in quantized_values]

# Print the first few binary codes for inspection
print("First few binary codes:")
for val, binary in zip(quantized_values[:10], binary_codes[:10]):
    print(f"Quantized value: {val:.2f} -> Binary code: {binary}")

# Plot the quantized signal with binary codes as labels and modify the y-axis grid
plt.figure(figsize=(10, 4))
plt.plot(t_analog, y_analog, label='Analog Signal')
plt.stem(t_sampled, quantized_values, 'r', markerfmt='ro', basefmt='r-', label='Quantized Signal')

# Add binary code labels to every point
for t, q, b in zip(t_sampled, quantized_values, binary_codes):
    plt.text(t, q, b, fontsize=9, ha='center', color='blue')

# Set y-axis grid to match quantized amplitude levels
quantized_levels = np.linspace(-1, 1, num_levels)
plt.yticks(quantized_levels)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

plt.title('Quantized Signal with Binary Codes')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
````
