````python
import numpy as np
np.set_printoptions(suppress=True)
import matplotlib.pyplot as plt
````

- `import numpy as np`: Mengimpor perpustakaan `numpy` dan memberi alias `np`.
- `np.set_printoptions(suppress=True)`: Mengatur opsi tampilan `numpy` untuk menekan penggunaan notasi ilmiah saat mencetak array.
- `import matplotlib.pyplot as plt`: Mengimpor perpustakaan `matplotlib.pyplot` dan memberi alias `plt`.

## 1.1 Penggabungan Sinyal

````python
x = np.arange(0, 1, 0.0001)
print(x)
````

- `x = np.arange(0, 1, 0.0001)`: Membuat array `x` yang berisi nilai dari 0 hingga 1 dengan langkah 0.0001.
- `print(x)`: Mencetak nilai-nilai dalam array `x`.

````python
y1 = 3 + 2*np.cos(2 * np.pi * 4 * x * 4 + 0)
plt.figure(figsize=(17, 3))
plt.plot(x, y1); plt.show()
````

- `y1 = 3 + 2*np.cos(2 * np.pi * 4 * x * 4 + 0)`: Menghasilkan sinyal gelombang cosinus `y1` dengan amplitudo 2, frekuensi 4, dan ditambah offset 3.
- `plt.figure(figsize=(17, 3))`: Membuat gambar baru dengan ukuran 17x3 inci.
- `plt.plot(x, y1); plt.show()`: Memplot sinyal `y1` terhadap `x` dan menampilkan plot.

````python
y2 = 0 + 1 * np.cos(2 * np.pi * 5 * x + np.pi/6)
y3 = -1 + 5 * np.cos(2 * np.pi * 7 * x + np.pi/4)

fig, ax = plt.subplots(3, 1, figsize=(17, 6))
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)
plt.show()
````

- `y2 = 0 + 1 * np.cos(2 * np.pi * 5 * x + np.pi/6)`: Menghasilkan sinyal gelombang cosinus `y2` dengan amplitudo 1, frekuensi 5, dan pergeseran fase $\pi6$.
- `y3 = -1 + 5 * np.cos(2 * np.pi * 7 * x + np.pi/4)`: Menghasilkan sinyal gelombang cosinus `y3` dengan amplitudo 5, frekuensi 7, dan pergeseran fase $\pi4$.

- `fig, ax = plt.subplots(3, 1, figsize=(17, 6))`: Membuat gambar dengan 3 subplot dalam satu kolom dan ukuran 17x6 inci.
- `ax[0].plot(x, y1)`: Memplot sinyal `y1` pada subplot pertama.
- `ax[1].plot(x, y2)`: Memplot sinyal `y2` pada subplot kedua.
- `ax[2].plot(x, y3)`: Memplot sinyal `y3` pada subplot ketiga.
- `plt.show()`: Menampilkan plot.

````python
y123 = y1 + y2 + y3
plt.figure(figsize=(17, 3))
plt.plot(x, y123); plt.show()
````

- `y123 = y1 + y2 + y3`: Menghasilkan sinyal baru `y123` yang merupakan penjumlahan dari sinyal `y1`, `y2`, dan `y3`.
- `plt.figure(figsize=(17, 3))`: Membuat gambar baru dengan ukuran 17x3 inci.
- `plt.plot(x, y123); plt.show()`: Memplot sinyal `y123` terhadap `x` dan menampilkan plot.

## 1.2 Sine wave vs Cosine wave

````python
x = np.linspace(0, 2 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)
````

- `x = np.linspace(0, 2 * np.pi, 1000)`: Membuat array `x` yang berisi 1000 titik dari 0 hingga \(2\pi\).
- `y_sin = np.sin(x)`: Menghasilkan sinyal gelombang sinus `y_sin` berdasarkan array `x`.
- `y_cos = np.cos(x)`: Menghasilkan sinyal gelombang cosinus `y_cos` berdasarkan array `x`.

````python
plt.figure(figsize=(10, 4))
plt.plot(x, y_sin, label='Sine Wave')
plt.plot(x, y_cos, label='Cosine Wave')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(np.pi/2, color='red', linestyle='--', label='/2')
plt.axvline(3*np.pi/2, color='blue', linestyle='--', label='3/2')
plt.legend()
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
````

- `plt.figure(figsize=(10, 4))`: Membuat gambar baru dengan ukuran 10x4 inci.
- `plt.plot(x, y_sin, label='Sine Wave')`: Memplot sinyal `y_sin` terhadap `x` dengan label 'Sine Wave'.
- `plt.plot(x, y_cos, label='Cosine Wave')`: Memplot sinyal `y_cos` terhadap `x` dengan label 'Cosine Wave'.
- `plt.axhline(0, color='black', linewidth=0.5)`: Menambahkan garis horizontal pada y=0 dengan warna hitam dan lebar garis 0.5.
- `plt.axvline(np.pi/2, color='red', linestyle='--', label='/2')`: Menambahkan garis vertikal pada x=\(\pi/2\) dengan warna merah, gaya garis putus-putus, dan label '/2'.
- `plt.axvline(3*np.pi/2, color='blue', linestyle='--', label='3/2')`: Menambahkan garis vertikal pada x=\(3\pi/2\) dengan warna biru, gaya garis putus-putus, dan label '3/2'.
- `plt.legend()`: Menampilkan legenda untuk plot.
- `plt.title('Sine and Cosine Waves')`: Menambahkan judul 'Sine and Cosine Waves' pada plot.
- `plt.xlabel('x')`: Menambahkan label 'x' pada sumbu x.
- `plt.ylabel('Amplitude')`: Menambahkan label 'Amplitude' pada sumbu y.
- `plt.grid(True)`: Menambahkan grid pada plot.
- `plt.show()`: Menampilkan plot.
