Pada notebook ini, kita akan belajar bagaimana cara bekerja dengan sinyal dalam Python. Sinyal adalah data yang berubah seiring waktu. Sinyal dapat berupa suara, gambar, atau data lain yang berubah seiring waktu. Kita akan belajar bagaimana cara membuat sinyal dan memvisualisasikannya.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## Memahami persamaan sinyal

### Sinyal Sinusoidal

**Fungsi Sinyal Sinusoidal**

$$
\text{signal} = \sin \theta
$$

- Fungsi sinus adalah fungsi matematika yang menghasilkan nilai antara -1 dan 1. Fungsi ini sering digunakan untuk menggambarkan gelombang periodik, seperti gelombang suara atau gelombang cahaya.
(\theta):

**Fungsi `theta`**

$$
\theta = 2 \pi f t
$$

**Fungsi Sinyal Sinusoidal**

$$
\text{signal} = \sin (2 \pi f t)
$$

**Penjelasan**

1. **Sinyal**:
   - Dalam konteks ini, sinyal adalah data yang berubah seiring waktu. Contoh sinyal termasuk suara, gambar, atau data sensor.

2. **Fungsi Sinus (\(\sin\))**:
   - Fungsi sinus adalah fungsi matematika yang menghasilkan nilai antara -1 dan 1. Fungsi ini sering digunakan untuk menggambarkan gelombang periodik, seperti gelombang suara atau gelombang cahaya.

3. **\(\theta\)**:
   - \(\theta\) adalah variabel independen yang biasanya mewakili sudut dalam radian. Dalam konteks sinyal, \(\theta\) sering kali terkait dengan waktu atau frekuensi.

Pertama-tama kita buat time_axis (sumbu waktu / sumbu x)

````python
time_axis = np.linspace(0, 1, 1000)
# print(time_axis)
````

Kemudiannya kita buat sinyal sinusoidal dengan frekuensi 5 Hz

````python
frekuensi = 5
signal = np.sin(2 * np.pi * frekuensi * time_axis)
# print(signal)
````

Mari kita visualisasikan sinyal sinusoidal ini.

````python
plt.figure(figsize=(8, 3))
plt.plot(time_axis, signal)

# supaya lebih rapi, kita lakukan hal-hal berikut ini
plt.title("Sinyal Sinusoidal")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
````

### Membuktikan Representasi Sinyal Kompleks dengan Python

**Definisi Sinyal Kompleks:**
Sinyal kompleks adalah sinyal yang memiliki bagian real dan imajiner. Sinyal kompleks dapat direpresentasikan dalam bentuk:

$$
\text{complex-signal} = e^{j \theta}
$$

$$
\text{complex-signal} = e^{j 2 \pi f t}
$$

````python
time_axis = np.linspace(0, 1, 1000)
frekuensi = 5
complex_signal = np.exp(1j * 2 * np.pi * frekuensi * time_axis)
print(f"contoh sinyal complex element ke-3: {complex_signal[3]}")
````

Complex signal terdiri dari:
- Bagian real: $\cos(2 \pi f t)$
- dan Bagian imajiner: $\sin(2 \pi f t)$

````python
bagian_real = np.real(complex_signal)
bagian_imaginer = np.imag(complex_signal)

print(f"Bagian Real: {bagian_real[0:10]}")
print(f"Bagian Imaginer: {bagian_imaginer[0:10]}")
````

- `np.real` adalah fungsi yang digunakan untuk mendapatkan bagian real dari sinyal kompleks.
- `np.imag` adalah fungsi yang digunakan untuk mendapatkan bagian imajiner dari sinyal kompleks.

**Memvisualisasikan Sinyal Kompleks**

````python
# plot in subplots
ax, fig = plt.subplots(2, 1, figsize=(10, 4))
fig[0].plot(time_axis, bagian_real)
fig[0].set_title("Bagian Real (Cosinus)")
fig[0].grid(True)
fig[0].set_xlabel("Waktu (s)")
fig[0].set_ylabel("Amplitudo")
fig[1].plot(time_axis, bagian_imaginer)
fig[1].set_title("Bagian Imaginer (Sinus)")
fig[1].grid(True)
fig[1].set_xlabel("Waktu (s)")
fig[1].set_ylabel("Amplitudo")
plt.tight_layout()
plt.show()
````

Dapat dibuktikan bahwa:
- Bagian real dari sinyal kompleks berupa sinyal cosinus. (Terlihat sinyal dimulai dari 1)
- Bagian imajiner dari sinyal kompleks berupa sinyal sinus. (Terlihat sinyal dimulai dari 0)

**Mencoba untuk diplot dalam satu grafik**

````python
plt.figure(figsize=(10, 4))
plt.plot(time_axis, bagian_real, label="Bagian Real (Cosinus)")
plt.plot(time_axis, bagian_imaginer, label="Bagian Imaginer (Sinus)")
plt.title("Bagian Real dan Imaginer")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
````

## Bentuk Polar Sinyal

Bentuk polar adalah bagaimana sinyal dapat dilihat berdasarkan magnitudo dan fase (phase angle)

### Melihat perubahan magnitudo

Kita akan melihat perubahan magnitudo dari `complex-signal`

````python
magnitude = np.abs(complex_signal)

plt.figure(figsize=(10, 2))
plt.plot(time_axis, magnitude)
plt.title("Magnitude Sinyal")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
````

### Melihat perubahan fase


````python
phase = np.angle(complex_signal)

plt.figure(figsize=(10, 2))
plt.plot(time_axis, complex_signal)
plt.plot(time_axis, phase)
plt.title("Phase Sinyal")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
````

## Properti Sinyal (Dasar)

Memahami lebih detail tentang parameter-parameter penyusun sinyal
1. **Sampling Rate / Frekuensi Sampling (`fs`)**: Ini adalah jumlah sampel yang diambil per detik dan diukur dalam Hertz (Hz). Sampling rate menentukan resolusi sinyal dalam domain waktu.
   
2. **Vektor Waktu (`t`)**: Ini merepresentasikan titik-titik waktu diskrit di mana sinyal diambil sampelnya. Biasanya merupakan vektor yang terdistribusi secara linier.

3. **Frekuensi (`f`)**: Ini adalah jumlah siklus per detik (Hz) untuk sinyal periodik seperti gelombang sinus. Di SMP kita belajar, bahwa satu siklus adalah satu putaran penuh dari gelombang sinus. Atau dapat dilihat sebagai satu bukit dan satu lembah dari gelombang sinus.

4. **Amplitudo (`A`)**: Ini adalah nilai maksimum dari sinyal.

````python
fs = 100 # Hz
time_axis = np.linspace(0, 1, fs)
frekuensi = 5
amplitudo = 2
signal = amplitudo * np.sin(2 * np.pi * frekuensi * time_axis)

plt.figure(figsize=(19, 4))
plt.plot(time_axis, signal)

# supaya lebih rapi, kita lakukan hal-hal berikut ini
plt.title("Sinyal Sinusoidal")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
````

Sehingga, rumus sinyal sinusoidal, ketika bertemu dengan parameter-parameter ini, menjadi:

$$
\text{signal} = A \cdot \sin (2 \pi f t)
$$

### Sinyal Square Wave (Kotak)

- Kita akan gunakan library `scipy.signal` untuk melihat sinyal kotak

````python
from scipy import signal
````

````python
fs = 100 # Hz
time_axis = np.linspace(0, 1, fs)
amplitudo = 2
frekuensi = 5

signal_square = amplitudo * signal.square(2 * np.pi * frekuensi * time_axis)
````

```python
signal_square = amplitudo * signal.square(2 * np.pi * frekuensi * time_axis)
```

Sinyal square dapat ditulis sebagai persamaan:

$$
\text{signal} = A \cdot \text{square} (\theta)
$$

$$
\text{signal} = A \cdot \text{square}(2 \pi f t)
$$

````python
plt.figure(figsize=(10, 2))
plt.plot(time_axis, signal_square)
plt.title("Sinyal Square")
# supaya lebih rapi, kita lakukan hal-hal berikut ini
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.tight_layout()
plt.show()
````

### Sinyal Sawtooth Wave (Gergaji)

Sinyal square dapat ditulis sebagai persamaan:

$$
\text{signal} = A \cdot \text{sawtooth} (\theta)
$$

$$
\text{signal} = A \cdot \text{sawtooth}(2 \pi f t)
$$

````python
fs = 100 # Hz
time_axis = np.linspace(0, 1, fs)
amplitudo = 2
frekuensi = 5

signal_sawtooth = amplitudo * signal.sawtooth(2 * np.pi * frekuensi * time_axis)
````

````python
plt.figure(figsize=(10, 2))
plt.plot(time_axis, signal_sawtooth)
plt.title("Sinyal Sawtooth")
# supaya lebih rapi, kita lakukan hal-hal berikut ini
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.tight_layout()
plt.show()
````

### Sinyal Noise (Acak)

Sinyal noise merupakan sinyal yang tidak memiliki pola tertentu. Sinyal ini sering digunakan untuk merepresentasikan data yang tidak teratur atau tidak terduga.

> Agar noise yang dihasilkan tetap sama ketika menjalankan kode ini, kita perlu menetapkan nilai `random_state` pada `np.random.RandomState`.

````python
fs = 100 # Hz
time_axis = np.linspace(0, 1, fs)

np.random.seed(1945)
sinyal_noise = np.random.normal(0, 1, fs)
````

````python
plt.figure(figsize=(10, 2))
plt.plot(time_axis, sinyal_noise)
plt.title("Sinyal Noise")
# supaya lebih rapi, kita lakukan hal-hal berikut ini
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.tight_layout()
plt.show()
````

### Sinyal Komposit (Penjumlahan Dua buah Sinyal atau lebih)

Sinyal komposit adalah sebuah sinyal yang merupakan gabungan dari beberapa sinyal lain (compose+it). Sinyal ini sering digunakan untuk merepresentasikan data yang kompleks atau data yang terdiri dari beberapa komponen.

Mari membuat sinyal komposit:
- Pertama-tama buat dua buah sinyal sinusoidal dengan frekuensi yang berbeda.
- Kemudian, gabungkan kedua sinyal tersebut untuk membuat sinyal komposit.

````python
fs = 100 # Hz
time_axis = np.linspace(0, 1, fs)

frekuensi_1 = 3
frekuensi_2 = 9
amplitudo_1 = 2
amplitudo_2 = 4

signal_1 = amplitudo_1 * np.sin(2 * np.pi * frekuensi_1 * time_axis)
signal_2 = amplitudo_2 * np.sin(2 * np.pi * frekuensi_2 * time_axis)

signal_composite = signal_1 + signal_2
````

````python
ax, fig = plt.subplots(3, 1, figsize=(10, 6))
fig[0].plot(time_axis, signal_1)
fig[0].set_title("Sinyal 1")
fig[1].plot(time_axis, signal_2)
fig[1].set_title("Sinyal 2")
fig[2].plot(time_axis, signal_composite)
fig[2].set_title("Sinyal Composite")
plt.tight_layout()
plt.show()
````

Jika plotnya digabung

````python
plt.figure(figsize=(10, 2))
plt.plot(time_axis, signal_composite)
plt.plot(time_axis, signal_1)
plt.plot(time_axis, signal_2)
plt.title("Sinyal Composite")
plt.legend(["Composite", "Signal 1", "Signal 2"])
plt.tight_layout()
plt.show()
````
